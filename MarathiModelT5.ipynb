```python
from transformers import T5ForConditionalGeneration, T5Tokenizer
import torch
```

```python
model_name = "google/mt5-small"
tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)
```

```python
def simplify_text(text):
    input_text = f"simplify: {text}"
    input_ids = tokenizer.encode(input_text, return_tensors="pt")

    outputs = model.generate(input_ids, max_length=200, num_beams=4, early_stopping=True)
    simplified_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

    return simplified_text

complex_text = "आपले जीवन अधिक सुलभ करण्यासाठी वैज्ञानिक तंत्रज्ञानाचा वापर करतात."
simplified_text = simplify_text(complex_text)
print("Original:", complex_text)
print("Simplified:", simplified_text)
```

```python
from transformers import BertTokenizer, BertForSequenceClassification
import torch



model_name = 'nlptown/bert-base-multilingual-uncased-sentiment'

tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertForSequenceClassification.from_pretrained(model_name)

def simplify_text(text):
    inputs = tokenizer(text, return_tensors='pt', truncation=True, padding=True)

    outputs = model(**inputs)
    scores = outputs[0][0].detach().numpy()


    sentences = text.split('. ')


    sentence_scores = {sentence: score for sentence, score in zip(sentences, scores)}
    sorted_sentences = sorted(sentence_scores, key=sentence_scores.get, reverse=True)
    simplified_text = ' '.join(sorted_sentences[:2])

    return simplified_text

complex_text = "आपले जीवन अधिक सुलभ करण्यासाठी वैज्ञानिक तंत्रज्ञानाचा वापर करतात. हे तंत्रज्ञान अत्यंत प्रभावी आहे."
simplified_text = simplify_text(complex_text)
print("Original:", complex_text)
print("Simplified:", simplified_text)
```
